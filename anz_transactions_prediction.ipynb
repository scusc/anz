importing packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('ANZ synthesised transaction dataset.csv')
data
```

```python
data.columns
```

```python
data[data['txn_description'] == 'PAY/SALARY']
```

```python
salary = data[data['txn_description'] == 'PAY/SALARY'].groupby('customer_id').mean()
salary
```

```python
ann_sal =[]
for customer_id in data['customer_id']:
    ann_sal.append(int(salary.loc[customer_id]['amount']))
data['annual_salary'] = ann_sal
```

```python
len(ann_sal)
```

```python
data
```

```python
salaries = data[data['txn_description'] == 'PAY/SALARY']
salaries
```

```python
salaries = salaries.filter(['gender','balance','age','annual_salary'])
salaries
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df = salaries[['gender']].apply(le.fit_transform)
```

```python
salaries['gender'] = df
```

```python
salaries
```

```python
X = salaries.drop('annual_salary',axis = 1)
y = salaries['annual_salary']

from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.3)
```


linear regression model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train , y_train)
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

acc=mean_squared_error(y_test,y_pred,squared=False)
acc
```


decision tree model

```python
from sklearn.tree import DecisionTreeRegressor

dr = DecisionTreeRegressor()
dr.fit(X_train , y_train)
y_pred1 = dr.predict(X_test)
```

```python
acc1=mean_squared_error(y_test,y_pred1,squared=False)
acc1
```
